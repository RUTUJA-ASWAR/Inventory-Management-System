```python
#Read data from JSON
import json
fd = open("record.json",'r')
r=fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

```python
#Adding products to inventory
prod_id = str(input("Enter product id :"))
made_in = str(input("Enter the country of origin of product :"))
mfg_date= str(input("Enter the manufacturing date of product"))
name = str(input("Enter name of the product :"))
pr = int(input("Enter the price of product :"))
qn = int(input("Enter quantity of product :"))

record[prod_id] = {"name": name, 'pr': pr, 'qn':qn, "made_in": made_in, "mfg_date": mfg_date} 

js=json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()

print("New product is successfully added to your inventory")
```

```python
#Purchasing product 
no=int(input("enter a no of products you want to purchase\n"))

for i in range(1,no+1):
   ui_prod=str(input("enter a product id : "))
   ui_quant=int(input("enter a quantity you want to purachse"))

   print("product: ",record[ui_prod]['name'])
   print("price :",record[ui_prod]['pr'])
   print("billing amount : ", record[ui_prod]['pr']* ui_quant)

   record[ui_prod]['qn']=record[ui_prod]['qn'] + ui_quant
```

```python
#Update inventory after purchase
js=json.dumps(record)

fd=open("record.json",'w')
fd.write(js)
fd.close()
```
