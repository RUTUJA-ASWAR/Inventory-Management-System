```python
import json

fd=open("record.json",'r')
r=fd.read()
fd.close()

record=json.loads(r)

record
```

```python
import time
sales={}

no=int(input("Enter the number of products customer want to puchase : "))


for i in range(1,no+1):
   no= int(input("enter a key number : "))
   ui_prod=str(input("enter a product id: "))
   ui_quant=int(input("enter a quantity : "))
  
   if record[ui_prod]['qn'] >= ui_quant :
       print("name: ",record[ui_prod]['name'])
       print ("Billling amount : ",record[ui_prod]['pr']*ui_quant)

       sales[no]={'prod': ui_prod, 'qn': ui_quant, 'amount': record[ui_prod]['pr']*ui_quant, 'name': record[ui_prod]['name'], 'time': time.ctime()}
   
    
       record[ui_prod]['qn']=record[ui_prod]['qn'] - ui_quant
   else:
      print("sorry,such qunatity is not avialable\n") 


print("total sales made")
```

```python
js=json.dumps(sales)

fd=open("sales.json",'w')
r=fd.write(js)
fd.close()
```
